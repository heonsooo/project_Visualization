```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
```

```python
# txt파일은 2021-02-08 22:03 기준 랭킹 뉴스 
# url = https://news.naver.com/main/ranking/read.nhn?mode=LSD&mid=shm&sid1=001&oid=243&aid=0000011772&rankingType=RANKING
with open('hello.txt',encoding='utf-8') as f:
    lines = f.readlines()
    string = ''.join([line.rstrip() for line in lines])
print(type(string))
```

```python
wordcloud = WordCloud(font_path='./ttf/NanumGothic.ttf',
                     background_color='white',
                      colormap = 'Accent_r',# 폰트 색깔 정하기
                      width = 800, height = 800).generate(string)
```

```python
plt.figure(figsize=(10,10)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show() 
# plt.savefig()
```

```python
# 불용어 제거 
from wordcloud import STOPWORDS
 
stopwords = set(STOPWORDS) 
stopwords.add('시장에서') 

# wordcloud = WordCloud(font_path='./ttf/NanumGothic.ttf',stopwords=stopwords,background_color='white').generate(string)
wordcloud = WordCloud(font_path='./ttf/NanumGothic.ttf',
                      stopwords=stopwords,
                     background_color='white',
                      colormap = 'Accent_r',# 폰트 색깔 정하기
                      width = 800, height = 800).generate(string)
```

```python
fig = plt.figure(figsize=(10,10)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show() 
```

```python
fig.savefig('news_worldcloud.png')
```
