```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
```

```python
# txt파일은 2020 3분기 보고서 
# url = 'http://dart.fss.or.kr/dsaf001/main.do?rcpNo=20201116001607'
with open('./data/emart.txt',encoding='utf-8') as f:
    lines = f.readlines()
    string = ' '.join([line.rstrip() for line in lines])
print(type(string))
```

```python
wordcloud = WordCloud(font_path='./ttf/NanumGothic.ttf',
                     background_color='white',
                      colormap = 'Accent_r',# 폰트 색깔 정하기
                      width = 800, height = 800).generate(string)
```

```python
plt.figure(figsize=(10,10)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show() 
# plt.savefig()
```

```python
# 불용어 제거 
from wordcloud import STOPWORDS
 
stopwords = set(STOPWORDS)
words = ['있습니다', '있으며', '따른','등의','다양한','따라','있는','등을','통해','대한','또한']
for i in words:
    stopwords.add(i) 

wordcloud = WordCloud(font_path='./ttf/NanumGothic.ttf',
                      stopwords=stopwords,
                     background_color='white',
                      colormap = 'Accent_r',# 폰트 색깔 정하기
                      width = 800, height = 800).generate(string)
```

```python
fig = plt.figure(figsize=(10,10)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show() 
```

```python
fig.savefig('emart_사업내용_wordcloud_1.png')
```
